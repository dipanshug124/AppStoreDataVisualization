```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=pd.read_csv('appstore_games.csv')
```

```python
a.describe()
```

###### Data Cleaning

```python
a.drop(['URL','ID','Icon URL','Subtitle','Description','Developer','Primary Genre'],axis=1,inplace=True)
```

```python
a.columns
```

```python
#Removing NULL values
a=a[pd.notnull(a['Average User Rating'])]
```

```python
#converting date strings to date time object
a['Original Release Date']=pd.to_datetime(a['Original Release Date'])
a['Current Version Release Date']=pd.to_datetime(a['Current Version Release Date'])
```

```python
#add column for gap between release and update
a['Update Gap']=a['Current Version Release Date'] - a['Original Release Date']
```

```python
#remove rating less than 200 user count and update gap less than 6 months
baised=(a[a['User Rating Count'] < 200].index) & (a[a['Update Gap'] < datetime.timedelta(days=180)].index)
a.drop(baised,inplace =True)
```

```python
#remove genre tags without meaningfull insight
a.Genres=a.Genres.str.replace('Games','').str.replace('Entertainment','').str.replace(',','')
a.Genres=a.Genres.str.split(' ').map(lambda x: ' '.join(sorted(x)))
a.Genres=a.Genres.str.strip()
```

```python
#Remove non essential Genres
non_ess=a[~a.Genres.str.contains('Puzzel') & \
         ~a.Genres.str.contains('Board') & \
         ~a.Genres.str.contains('Adventure') & \
          ~a.Genres.str.contains('Role') & \
         ~a.Genres.str.contains('Role Playing') & \
         ~a.Genres.str.contains('Action') & \
         ~a.Genres.str.contains('Family') & \
         ~a.Genres.str.contains('Education')].index
a.drop(non_ess, inplace=True)
```

```python
#Grouping strings in Genres
a.loc[a.Genres.str.contains('Puzzel'),'Genres']='Puzzel'
a.loc[a.Genres.str.contains('Board'),'Genres']='Puzzel'
a.loc[a.Genres.str.contains('Adventure'),'Genres']='Adventure'
a.loc[a.Genres.str.contains('Role'),'Genres']='Adventure'
a.loc[a.Genres.str.contains('Role Playing'),'Genres']='Adventure'
a.loc[a.Genres.str.contains('Action'),'Genres']='Action'
a.loc[a.Genres.str.contains('Family'),'Genres']='Family'
a.loc[a.Genres.str.contains('Education'),'Genres']='Family'
```

###### Data Visualization

```python
# Number of Games w.r.t. Types of Genres

%matplotlib inline 
import warnings
warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
f,axes= plt.subplots(2,1,figsize=(8,8))

#Histogram
x=['Puzzel','Action','Adventure','Family']
y=[a.Genres[(a.Genres=='Puzzel')].count() ,
   a.Genres[(a.Genres=='Adventure')].count() ,
   a.Genres[(a.Genres=='Family')].count() ,
   a.Genres[(a.Genres=='Action')].count()]

graph1=sns.barplot(x,y,palette='Accent',ax=axes[0])
graph1.set(xlabel='Genres', ylabel='No of Games')
for p in graph1.patches:
    graph1.annotate("%.f" % p.get_height() , (p.get_x() + p.get_width()/2., p.get_height()) ,
                    ha='center', va='center', fontsize=11, color='gray', 
                    xytext=(0,20),textcoords='offset points')
                    
        
#PieChart
z=[a.Genres[(a.Genres=='Puzzel')].count() ,
   a.Genres[(a.Genres=='Adventure')].count() ,
   a.Genres[(a.Genres=='Family')].count() ,
   a.Genres[(a.Genres=='Action')].count()]
q=['Puzzel','Action','Adventure','Family']

plt.pie(z, labels=q, startangle=90, autopct='%.1f%%')
plt.show()
plt.ioff()
```

```python
#Genres with higher user rating

x=['Puzzel','Action','Adventure','Family']
y=[a['Average User Rating'][(a.Genres=='Puzzel')].mean() ,
   a['Average User Rating'][(a.Genres=='Adventure')].mean() ,
   a['Average User Rating'][(a.Genres=='Family')].mean() ,
   a['Average User Rating'][(a.Genres=='Action')].mean()]

Genres_Graph=sns.barplot(x,y,palette='Accent')
Genres_Graph.set(xlabel='Genres', ylabel='Average user rating')
Genres_Graph.set(ylim=(3,5))

for p in Genres_Graph.patches:
    Genres_Graph.annotate("%.2f" % p.get_height() , (p.get_x() + p.get_width()/2., p.get_height()) ,
                    ha='center', va='center', fontsize=11, color='gray', 
                    xytext=(0,20),textcoords='offset points')
plt.ioff()
```

```python
#Genres v/s Size of Game

a['Size2'] = round(a['Size']/1000000,1)
Size_Less_250MB = a[a.Size2 <250]
Size_More_250MB = a[(a.Size2 >=250) & (a.Size2 <1000)]
Size_More_1GB = a[a.Size2 >=1000]


vis1= sns.stripplot(y='Size2' , x='Genres', data=a,\
                hue='Average User Rating',dodge=True, size=4)

plt.legend(title='User Rating(1-5)',bbox_to_anchor=(1.29, 1))

vis1.set(ylim=(0, 1000))
vis1.set(xlabel='User Rating (1-5)',ylabel='Game Size in MB')
plt.ioff()
```

```python
#Size of games v/s User Ratings

sns.set_style('darkgrid')
f, axes = plt.subplots (2,3, figsize=(18,12))



vis1 = sns.boxplot( data = Size_Less_250MB, x = 'Average User Rating', y = 'Size2',  ax=axes[1,0])
vis2 = sns.boxplot( data = Size_More_250MB, x = 'Average User Rating', y = 'Size2', ax=axes[1,1])
vis3 = sns.boxplot( data = Size_More_1GB, x = 'Average User Rating', y = 'Size2', ax=axes[1,2],)
vis4 = sns.distplot(Size_Less_250MB.Size2, bins= 10, kde=False,ax=axes[0,0])
vis5 = sns.distplot(Size_More_250MB.Size2, bins= 10, kde=False,ax=axes[0,1])
vis6 = sns.distplot(Size_More_1GB.Size2, bins= 10, kde=False,ax=axes[0,2])

vis1.set(xlabel='User Rating (1-5)',ylabel='Game Size in MB')
vis2.set(xlabel='User Rating (1-5)',ylabel='Game Size in MB')
vis3.set(xlabel='User Rating (1-5)',ylabel='Game Size in MB')
vis4.set(xlabel='Game Size in MB',ylabel='Number of Games')
vis5.set(xlabel='Game Size in MB',ylabel='Number of Games')
vis6.set(xlabel='Game Size in MB',ylabel='Number of Games')

axes[0, 0].set_title('No. of Games Below 250MB')
axes[0, 1].set_title('No. of Games Between 250MB and 1GB')
axes[0, 2].set_title('No. of Games Above 1GB')
axes[1, 0].set_title('Games Below 250MB vs User Rating')
axes[1, 1].set_title('Games Between 250MB and 1GB vs User Rating')
axes[1, 2].set_title('Games Above 1GB vs User Rating')
vis3.set(xlim=(-4, 5.5))

vis4.set(ylim=(0, 800))
vis5.set(ylim=(0, 800))
vis6.set(ylim=(0, 800))

plt.ioff()
```

```python
# Variation in size of Game

a['Year'] = a['Original Release Date']
a['Year']=a['Year'].astype(str)
#extract year ony
for i in (list(a.index.values)):
    a['Year'][i]=a['Year'][i][:4]
    
g, axes = plt.subplots (1,2, figsize=(16,6))

vis1 = sns.lineplot(x=a.Year, y=a.Size2, data=a, ax=axes[0])
vis1.set(ylim=(0, 600))
vis2 = sns.lineplot(x=a.Year, y=a.Size2, hue=a.Genres, err_style=None, marker='o',ax=axes[1])
vis2.set(ylim=(0, 600))
plt.legend(bbox_to_anchor=(1, 1))

axes[0].set_title('Game size changes over 12 years',fontsize=15)
axes[1].set_title('Game size changes over 12 years by Genre',fontsize=15)

vis1.set(xlabel='Year',ylabel='Game Size in MB')
vis2.set(xlabel='Year',ylabel='Game Size in MB')

plt.ioff()
```

```python
# Average User Rating v/s Price

h, axes = plt.subplots (2,1, figsize=(8,10))

vis1 = sns.distplot(a.Price,kde=None,ax=axes[0])
vis2 = sns.stripplot(y=a.Price, x=a.Genres, data=a,ax=axes[1],hue=a['Average User Rating'], dodge=1)

vis1.set(xlabel='Price',ylabel='Number of Games')
vis2.set(xlabel='Genre',ylabel='Game Price in $')

axes[0].set_title('Game price distribution',fontsize=15)
axes[1].set_title('User Rating vs Price',fontsize=15)
plt.legend(title='User Rating(1-5)',bbox_to_anchor=(1.22, 1))

plt.ioff()
```

```python
# In-app purchase v/s Price

a2=a.copy()
import statistics
a2['In-app Purchases'] = a2['In-app Purchases'].str.replace(',', '') 
a2['In-app Purchases']=a2['In-app Purchases'].fillna(0)

for i in (list(a2.index.values)):
    if a2['In-app Purchases'][i] ==0:
        a2['In-app Purchases'][i]=0.01
    elif a2['In-app Purchases'][i] !=0:
        A=str.split(a2['In-app Purchases'][i])
        A=[float(i) for i in A]
        a2['In-app Purchases'][i]=statistics.mean(A)
vis1= sns.stripplot(y=a2['In-app Purchases'] , x=a2.Price, data=a2, hue=a2.Genres, dodge=True, size=3)

vis1.set(xlabel='Price in $',ylabel='Average in App Purchase in $')
vis1.set(xlim=(-0.7, 12))
plt.ioff()
```

```python
# Age rating

NAR = [a2['Age Rating'][(a2['Age Rating']=='4+')].count(),a2['Age Rating'][(a2['Age Rating']=='9+')].count(),\
     a2['Age Rating'][(a2['Age Rating']=='12+')].count(),a2['Age Rating'][(a2['Age Rating']=='17+')].count()]
AR = ['Age 4+','Age 9+','Age 12+','Age 17+']

plt.pie(NAR, labels=AR, startangle=90, autopct='%.1f%%')
plt.show()
```

```python
# Age Rating v/s User Rating

vis1 = sns.boxplot(x=a2['Age Rating'], y=a['Average User Rating'], data=a, hue=a.Genres, dodge=True)

vis1.set(xlabel='Age Rating',ylabel='User Rating (1-5)')
plt.legend(title='User Rating(1-5)',bbox_to_anchor=(1.28, 1))
plt.ioff()
```

* Family Games show lesser rating with size above 200MB.  

###### Game Size Analysis  
* Games below 250MB -> focus on size between 100MB to 150MB of contents.  
* Games between 250MB to 1GB -> focus on size between 500MB+ of contents. 

###### Release Date/ Update Factor  
* Almost 90% of the developers focuses on games below 1 Dollar and places In-App Purchase as their strategy for income

###### Game price and In-App Purchase Factor  
* A large number of Developers create free games.  
* Puzzel shows more no. of paid players compare to others.  

###### Age Rating Factor  
* Games with age rating 12+ gains more likes.
